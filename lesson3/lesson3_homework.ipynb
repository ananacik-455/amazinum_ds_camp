````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
````

# energy

````python
# Read and process initial data
energy = pd.read_excel('data/Energy Indicators.xls', skiprows=16)
energy = energy.iloc[:, 2:]
energy = energy.dropna()
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy = energy.reset_index(drop=True)
energy
````

````python
energy.info()
````

````python
energy = energy.replace({"...": np.NaN})
energy.iloc[:, 1:] = energy.iloc[:, 1:].astype(np.float64)
energy
````

````python
energy.loc[:, 'Energy Supply'] *= 1_000_000
energy.loc[:, 'Energy Supply'] = energy.loc[:, 'Energy Supply']
energy
````

````python
s = '0123456789()'
indexes = energy[energy.Country.str.endswith(tuple(str(x) for x in s))].index
energy[energy.Country.str.endswith(tuple(str(x) for x in s))]
````

````python
energy.Country = energy.Country.apply(lambda x: re.sub(r"[0-9]", '', x))
energy.Country = energy.Country.apply(lambda x: re.sub(r"\([^()]*\)", "", x))
energy.Country = energy.Country.apply(lambda x: x.strip())
energy.iloc[indexes]
````

````python
# country renaming
renaming_dict = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong",
    "Viet Nam": "Vietnam"
}
energy[energy.Country.isin(renaming_dict.keys())]
````

````python
energy.Country = energy.Country.replace(renaming_dict)
energy[energy.Country.isin(renaming_dict.values())]
````

## GDP

````python
GDP = pd.read_csv('data/world_bank.csv', skiprows=4)
GDP.head()
````

````python
rename = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}
GDP[GDP['Country Name'].isin(rename.keys())]
````

````python
GDP['Country Name'] = GDP['Country Name'].replace(rename)
GDP[GDP['Country Name'].isin(rename.values())]
````

````python
GDP['Country'] = GDP['Country Name']
GDP = GDP.drop(columns=['Country Name', 'Unnamed: 67'])
GDP
````

````python
ScimEn = pd.read_excel('data/scimagojr country rank 1996-2022.xlsx')
ScimEn.head(15)
````

Join the three datasets: Energy, GDP, and ScimEn into a new dataset (using the intersection of country names). Use only the 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15).

````python
columns = ['Country', 'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
def answer_one():
    joined = pd.merge(energy, GDP, on='Country')
    joined = pd.merge(joined, ScimEn, on='Country')
    joined = joined.sort_values('Rank')
    joined = joined.loc[:, columns]
    joined = joined.set_index('Country')
    return joined.head(15)
````

## Answer the following questions in the context of only the top 15 countries by Scimagojr Rank (aka the DataFrame returned by `answer_one()`)

### Question 2
What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This function should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

````python
def answer_two():
    Top15 = answer_one()
    return Top15.iloc[:, -10:].mean(axis=1).sort_values(ascending=False)
````

````python
answer_two()
````

### Question 3
By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

*This function should return a single number.*

````python
def answer_three():
    Top15 = answer_one()
    avg_gdp = answer_two()
    index_6th_avg = avg_gdp.index[5]
    return Top15.loc[index_6th_avg, '2015'] - Top15.loc[index_6th_avg, '2006']
````

````python
answer_three()
````

### Question 4

Create a new column that is the ratio of Self-Citations to Total Citations. 
What is the maximum value for this new column, and what country has the highest ratio?

*This function should return a tuple with the name of the country and the ratio.*

````python
def answer_four():
    Top15 = answer_one()
    Top15['Citations Ratio'] = Top15['Self-citations'] / Top15['Citations']
    Top15 = Top15.sort_values('Citations Ratio', ascending=False)
    country = Top15.index[0]
    value = Top15.loc[country, 'Citations Ratio']
    return (country, value)
````

````python
answer_four()
````

### Question 5

Create a column that estimates the population using Energy Supply and Energy Supply per capita. 
What is the third most populous country according to this estimate?

*This function should return a single string value.*

````python
def answer_five():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15 = Top15.sort_values('Population', ascending=False)
    return Top15.index[2]
````

````python
answer_five()
````

### Question 6
Create a column that estimates the number of citable documents per person. 
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*


````python
def answer_six():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable documents per person'] = Top15['Citable documents'] / Top15['Population']
    Top15 = Top15.sort_values('Citable documents per person', ascending=False)
    return Top15.corr()['Citable documents per person']['Energy Supply per Capita']
````

````python
answer_six()
````

### Question 7
Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
````

````python
def answer_seven():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Continent'] = Top15.index.to_series().apply(lambda x: ContinentDict[x])
    grouped = Top15.groupby('Continent').agg({'Population': ['size', 'sum', 'mean', 'std']})
    return grouped
````
